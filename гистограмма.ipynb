```python
import matplotlib.pyplot as plt
x = ['Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь', 'Январь', 'Февраль']
y = [150, 110, 80, 50, 20, 2]

plt.bar(x, y, label='Данные с календаря', alpha=0.1 )
plt.plot(x, y, color='red', marker='o', markersize=7)
plt.xlabel('Месяц года')
plt.ylabel('Кол-во человек')
plt.title('Количество людей на лекциях')
plt.legend(loc='upper right')
plt.show()
```
